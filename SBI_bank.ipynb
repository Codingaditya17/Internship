```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:/Users/91902/OneDrive/Desktop/internship/SBIN_Data.csv")
```

```python
df['rank'] = 0
```

```python
df.head()
```

```python
for index, row in df.iterrows():
    current_date = row['Date']
    current_time = row['Time']
    last_5_days_data = df[(df['Date'] < current_date) & 
                          (df['Time'] == current_time)].tail(5)
    rank = (last_5_days_data['Volume'] > row['Volume']).sum() + 1 
    df.at[index, 'rank'] = rank
print(df)
```
